```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('Wine.csv')
```

```python
df
```

```python
df.info()
```

```python
df.dtypes
```

```python
df.shape
```

```python
for col in df.columns:
    print(f"Column {col} value counts:")
    print(df[col].value_counts())
    print("\n")
```

```python
df.isnull().sum()
```

```python
features = df.columns.difference(['Customer_Segment'])
x = df.loc[:, features].values
y = df.loc[:, ['Customer_Segment']].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler() 
```

```python
x_train_std = sc.fit_transform(x_train)
x_test_std = sc.fit_transform(x_test)
```

```python
x_train_std
```

```python
x_test_std
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
x_train_transform = pca.fit_transform(x_train)
x_test_transform = pca.transform(x_test)

explained_variance = pca.explained_variance_ratio_
```

```python
explained_variance
```

```python
eigenvalues = pca.explained_variance_
print('eigenvalues of PCA ' ,eigenvalues)
```

```python
eigenvectors = pca.components_
print('eigenvectors of PCA' , eigenvectors) 
```

```python
principal_df = pd.DataFrame(data=x_train_transform, columns=['PC1', 'PC2'])
```

```python
# Concatenate with the target variable
final_df = pd.concat([principal_df, df[['Customer_Segment']]], axis=1)
```

```python
final_df
```

```python
final_df.head()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(final_df['PC1'] , final_df['PC2'] , c=final_df['Customer_Segment'] , cmap = 'CMRmap')
plt.title('PCA of wine dataset')
plt.xlabel('PCA_1')
plt.ylabel('PCA_2')
plt.colorbar()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
